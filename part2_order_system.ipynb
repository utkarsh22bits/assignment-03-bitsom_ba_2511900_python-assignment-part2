```python
# ============================================================
# Assignment - Part 2: Data Structures
# Theme: Restaurant Menu & Order Management System
# ============================================================

# ---- PROVIDED DATA ----
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# ============================================================
# TASK 1 - Explore the Menu
# ============================================================

print("=" * 50)
print("           TASK 1 - Explore the Menu")
print("=" * 50)

# -------------------------------
# 1. Print menu grouped by category
# -------------------------------

# First, get unique categories
categories = set()

for item in menu:
    categories.add(menu[item]["category"])

# Loop through each category
for category in categories:
    print(f"\n===== {category} =====")

    # Loop through menu items and filter by category
    for item_name, details in menu.items():
        if details["category"] == category:

            price = details["price"]
            availability = "Available" if details["available"] else "Unavailable"

            # simple formatting (not perfect but readable)
            print(f"{item_name:<15} ₹{price:.2f}   [{availability}]")

# -------------------------------
# 2. Dictionary stats
# -------------------------------

# Total number of items
total_items = len(menu)

# Total available items
available_count = 0
for item in menu.values():
    if item["available"]:
        available_count += 1

# Most expensive item
max_price = 0
expensive_item = ""

for name, details in menu.items():
    if details["price"] > max_price:
        max_price = details["price"]
        expensive_item = name

# Items under ₹150
cheap_items = []

for name, details in menu.items():
    if details["price"] < 150:
        cheap_items.append((name, details["price"]))

# -------------------------------
# Print results
# -------------------------------

print("\n--- Menu Summary ---")
print(f"Total items           : {total_items}")
print(f"Available items       : {available_count}")
print(f"Most expensive item   : {expensive_item} (₹{max_price})")

print("\nItems under ₹150:")
for item in cheap_items:
    print(f"{item[0]} - ₹{item[1]}")



# ============================================================
# TASK 2 - Cart Operations
# ============================================================

print("\n" + "=" * 50)
print("           TASK 2 - Cart Operations")
print("=" * 50)

# Start with empty cart
cart = []

# -------------------------------
# Helper functions
# -------------------------------

def print_cart():
    print("\nCurrent Cart:")
    if not cart:
        print("Cart is empty")
        return

    for item in cart:
        print(f"{item['item']} x{item['quantity']} (₹{item['price']})")


def add_to_cart(item_name, quantity):
    # Check if item exists
    if item_name not in menu:
        print(f"{item_name} not found in menu.")
        return

    # Check availability
    if not menu[item_name]["available"]:
        print(f"{item_name} is currently unavailable.")
        return

    # Check if already in cart
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += quantity
            print(f"Updated {item_name} quantity to {item['quantity']}")
            return

    # If not in cart, add new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f"Added {item_name} x{quantity} to cart")


def remove_from_cart(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            print(f"Removed {item_name} from cart")
            return

    print(f"{item_name} not found in cart.")


def update_quantity(item_name, new_quantity):
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] = new_quantity
            print(f"Updated {item_name} quantity to {new_quantity}")
            return

    print(f"{item_name} not found in cart.")


# -------------------------------
# Simulation steps
# -------------------------------

add_to_cart("Paneer Tikka", 2)
print_cart()

add_to_cart("Gulab Jamun", 1)
print_cart()

add_to_cart("Paneer Tikka", 1)  # should update to 3
print_cart()

add_to_cart("Mystery Burger", 1)  # doesn't exist
print_cart()

add_to_cart("Chicken Wings", 1)  # unavailable
print_cart()

remove_from_cart("Gulab Jamun")
print_cart()


# -------------------------------
# Final Order Summary
# -------------------------------

print("\n========== Order Summary ==========")

subtotal = 0

for item in cart:
    total_price = item["quantity"] * item["price"]
    subtotal += total_price

    print(f"{item['item']:<18} x{item['quantity']}    ₹{total_price:.2f}")

print("-" * 40)

gst = round(subtotal * 0.05, 2)
total = round(subtotal + gst, 2)

print(f"Subtotal:                ₹{subtotal:.2f}")
print(f"GST (5%):                ₹{gst:.2f}")
print(f"Total Payable:           ₹{total:.2f}")
print("=" * 40)

# ============================================================
# TASK 3 - Inventory Tracker with Deep Copy (6 marks)
# ============================================================

print("\n" + "=" * 50)
print("    TASK 3 - Inventory Tracker with Deep Copy")
print("=" * 50)

import copy

# -------------------------------
# 1. Deep copy inventory
# -------------------------------

# create a deep copy so original data is preserved
inventory_backup = copy.deepcopy(inventory)

# manually change one value to prove copy works
inventory["Paneer Tikka"]["stock"] = 5

print("After manual change:\n")

print("Inventory:")
print(inventory["Paneer Tikka"])

print("\nInventory Backup:")
print(inventory_backup["Paneer Tikka"])

# restore original inventory using backup
inventory = copy.deepcopy(inventory_backup)

print("\nInventory restored.\n")


# -------------------------------
# 2. Simulate order fulfilment
# -------------------------------

# using final cart from Task 2
for item in cart:
    item_name = item["item"]
    qty_needed = item["quantity"]

    # check if item exists in inventory (safe check)
    if item_name not in inventory:
        print(f"{item_name} not found in inventory.")
        continue

    stock_available = inventory[item_name]["stock"]

    # if enough stock, deduct normally
    if stock_available >= qty_needed:
        inventory[item_name]["stock"] -= qty_needed

    else:
        # if not enough stock, deduct whatever is available
        print(f"Warning: Only {stock_available} {item_name} available, cannot fulfill full order.")
        inventory[item_name]["stock"] = 0  # cannot go below 0


# -------------------------------
# 3. Reorder Alerts
# -------------------------------

print("\nReorder Alerts:")

for item_name, details in inventory.items():
    stock = details["stock"]
    reorder_level = details["reorder_level"]

    # trigger alert if stock is low
    if stock <= reorder_level:
        print(f"⚠ Reorder Alert: {item_name} — Only {stock} unit(s) left (reorder level: {reorder_level})")


# -------------------------------
# 4. Final comparison
# -------------------------------

print("\nFinal Inventory (after updates):")
print(inventory)

print("\nInventory Backup (unchanged):")
print(inventory_backup)


# ============================================================
# TASK 4 - Daily Sales Log Analysis (6 marks)
# ============================================================

print("\n" + "=" * 50)
print("       TASK 4 - Daily Sales Log Analysis")
print("=" * 50)

# -------------------------------
# 1. Total revenue per day
# -------------------------------

print("Revenue Per Day:")

daily_revenue = {}  # store revenue for each date

for date, orders in sales_log.items():
    total = 0

    # sum all order totals for that day
    for order in orders:
        total += order["total"]

    daily_revenue[date] = total
    print(f"{date} : ₹{total:.2f}")


# -------------------------------
# 2. Best-selling day
# -------------------------------

best_day = ""
max_revenue = 0

for date, revenue in daily_revenue.items():
    if revenue > max_revenue:
        max_revenue = revenue
        best_day = date

print(f"\nBest Selling Day: {best_day} (₹{max_revenue:.2f})")


# -------------------------------
# 3. Most ordered item
# -------------------------------

item_count = {}

# count how many times each item appears in orders
for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:

            if item not in item_count:
                item_count[item] = 1
            else:
                item_count[item] += 1

# find item with highest count
most_ordered_item = ""
max_count = 0

for item, count in item_count.items():
    if count > max_count:
        max_count = count
        most_ordered_item = item

print(f"Most Ordered Item: {most_ordered_item} ({max_count} times)")


# -------------------------------
# 4. Add new day and recompute
# -------------------------------

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("\nUpdated Revenue Per Day:")

daily_revenue = {}

for date, orders in sales_log.items():
    total = 0
    for order in orders:
        total += order["total"]

    daily_revenue[date] = total
    print(f"{date} : ₹{total:.2f}")

# recompute best day
best_day = ""
max_revenue = 0

for date, revenue in daily_revenue.items():
    if revenue > max_revenue:
        max_revenue = revenue
        best_day = date

print(f"\nNew Best Selling Day: {best_day} (₹{max_revenue:.2f})")


# -------------------------------
# 5. Numbered list of all orders
# -------------------------------

print("\nAll Orders:")

counter = 1

for date, orders in sales_log.items():
    for order in orders:

        # join item list into string
        items_str = ", ".join(order["items"])

        print(f"{counter}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_str}")

        counter += 1


```
